# Set-up

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline

df = pd.read_csv('penn_hr_data.csv')

# Print out all columns names in alphabetical order (just so they are easy to see).
for col in df.sort_index(axis = 1).columns:
    print(col)
```

# Histograms

```python
# Inspect what the data look like.
df.head()

# Make histograms (only for numeric columns that histograms would make sense for)
df_num = df.select_dtypes(include = ['float64', 'int64'])
df_num.drop(columns = ['IN_UNION', 'JOBS_ACTIVE', 'Job_Grade_Fiscal_Year', 'Manager_OfColor', 'HOME_DEPT_ORG', 
                      'HOME_SCHOOL_CTR', 'Manager_Years_Managed', 'Number_of_Reclass', 'Number_of_Sal_Adjustments',
                      'PRIMARY_APPT_JOB_CLASS', 'Percent_Career_Salary_Growth', 'Race_OfColor', 'Termination_Fiscal_Year',
                      'VOLUNTARY', 'Worker_ID'], inplace = True)

df_num.hist(figsize=(16, 20), bins=50, xlabelsize=8, ylabelsize=8);
```

```python
# Corresponding basic stats for these columns
df_num.describe()
```

# Correlation heatmap

```python
# Calculate correlation values between each variable.
correlations = df_num.corr()

# Create a heatmap
plt.figure(figsize=(14, 10))

sns.heatmap(correlations[(correlations >= 0.5) | (correlations <= -0.5)], 
            cmap='viridis', vmax=1.0, vmin=-1.0, linewidths=0.1,
            annot=True, annot_kws={"size": 8}, square=True);
```

## Cleaning the columns

#### Country of citizenship

```python
print(df['COUNTRY_OF_CITIZENSHIP'].unique())   # '91'... Is that a country? Need to inspect that row. 

# Countries with most number of employees
df['COUNTRY_OF_CITIZENSHIP'].value_counts()   
#       18236
# CH      383
# IN      200
# CA       58
# KS       48
# UK       34
# RS       25
# TW       24

df[df['COUNTRY_OF_CITIZENSHIP'] == '91']
```

#### Home city

```python
print(df['HOME_CITY'].value_counts())   # every entry has extra white spaces... remove it.
                                        # PHILA == Philadelphia?
    
df['HOME_CITY'] = df['HOME_CITY'].str.strip()   # gets rid of leading and trailing white space

wrong_philly = df['HOME_CITY'] == 'PHILA'   # Let's replace "PHILA" with PHILADELPHIA.
df['HOME_CITY'][wrong_philly] = 'PHILADELPHIA'


''' Cities with >= 100 workers '''
'''
PHILADELPHIA     9942
HAVERTOWN         267
CHERRY HILL       222
DREXEL HILL       213
UPPER DARBY       183
MEDIA             181
WYNNEWOOD         173
WILMINGTON        164
WEST CHESTER      162
LANSDOWNE         158
BRYN MAWR         136
ARDMORE           127
WALLINGFORD       125
SPRINGFIELD       116
GLENSIDE          100
'''
```

#### Visa

```python
print(df['VISA'].value_counts())
```

#### Visa expiration date

```python
df['VISA_EXPIRATION_DATE'].value_counts()   # A lot of visas seem to expire btw June-Aug... let's group by months.

df['VISA_EXPIRATION_DATE'].apply(lambda x: str(x).split('/')[0]).value_counts()   # count by month
df['VISA_EXPIRATION_DATE'].apply(lambda x: str(x)[-9:-5]).value_counts()   # count by year


# Some ppl's visa ends in 1991, 1998, and 2036.
df[df['VISA_EXPIRATION_DATE'].apply(lambda x: str(x)[-9:-5]) == '2036']   # who has 2036 visa?
```

```python
df.groupby('VISA').agg({'VISA': 'count', 
                        'CURRENT_AGE': ['mean','min', 'max'], 
                        'Race_OfColor': 'sum', 
                        'PRIMARY_APPT_SALARY': ['mean','min', 'max']})
```

#### Primary appointment salary

```python
df['PRIMARY_APPT_SALARY'].mean()   # 61168.54
df['PRIMARY_APPT_SALARY'].median()   # 51500
df['PRIMARY_APPT_SALARY'].max()   # 393666
df['PRIMARY_APPT_SALARY'].min()   # 8824.4
```

```python
df[df['PRIMARY_APPT_SALARY'] >= 300000]   # 6 total. All white. 4/6 are male. 5/6 are in their 50s.
df[df['PRIMARY_APPT_SALARY'] >= 200000]['Race_OfColor'].sum()   # 19/147 are of color
df[df['PRIMARY_APPT_SALARY'] >= 200000].groupby('SEX').agg({'SEX':'count', 
                                                            'Race_OfColor':'sum', 
                                                            'CURRENT_AGE':['min', 'max', 'mean']})
```
